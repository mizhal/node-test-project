```python
"""
Demo of scatter plot on a polar axis.

Size increases radially in this example and color increases with angle (just to
verify the symbols are being scattered correctly).
"""

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt


N = 150
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
area = 200 * r**2 * np.random.rand(N)
colors = theta

ax = plt.subplot(111, polar=True)
c = plt.scatter(theta, r, c=colors, s=area, cmap=plt.cm.hsv)
c.set_alpha(0.75)

plt.show()
```

```python
''' Prueba con un solo error '''

%matplotlib inline

import math
error_center = [50,50]
R = 25

def vec2dDistance(vec1, vec2):
    d_x = float(vec2[0] - vec1[0])
    d_y = float(vec2[1] - vec1[1])
    return math.sqrt(d_x**2 + d_y**2)

print("Valor de e = %s" % math.e)

K = (math.e - 1) / R
decay_function = lambda x: 1 - math.log(1 + K*x)

print("Valor de decay en %s = %s" % (0, decay_function(0)))
print("Valor de decay en %s = %s" % (R, decay_function(R)))

mat = [[j for j in range(0,100)] for i in range(0, 100)]

for i in range(0,100):
    for j in range(0,100):
        r = vec2dDistance([i, j], error_center)
        mat[i][j] = 255 * decay_function(r)
        if mat[i][j] < 0:
            mat[i][j] = 0
        
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.collections import PatchCollection

figure, ax = plt.subplots()
ax.imshow(mat)

### Mostrar el perfil de la funcion decay
from pylab import arange

figure, ax2 = plt.subplots()

def zero_cap(v):
    if v < 0:
        return 0
    else:
        return v

x_line = arange(0,100,1)
y = [zero_cap(decay_function(x)) for x in x_line]
decay_curve, = ax2.plot(x_line, y, label="decay(r)")
plt.legend(handles = [decay_curve])
```
